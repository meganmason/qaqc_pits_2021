```python
import glob
import os
from pathlib import Path
import datetime
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="whitegrid")
sns.set_context("talk", font_scale=1.1)
```

```python
path = '/Users/meganmason491/Documents/snowex/2021/TS-sites/SNEX_2021_SiteVisits_withPhotos.xlsx'
```

```python
d = pd.read_excel(path, sheet_name='Summary')
```

```python
dloc = d[['Location', 'Pit Count']][:7]
```

```python
plt.figure(figsize=(10,8))
ax = sns.barplot(x="Location", y='Pit Count', data=dloc)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha="right");
plt.tight_layout()
plt.savefig("figs/pitcount-byLoc.png", format='png',dpi=150)
```

```python
dsite = d.iloc[:, :4]
dsite.head()
```

```python
plt.figure(figsize=(16,8))
ax = sns.barplot(x="Site", y='PitCount', hue='TS-Location', data=dsite, dodge=False) #dodge avoids skinny bars...
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right');
ax.set(yticks= range(0,20,2));
plt.legend(bbox_to_anchor=(1.01, 1),borderaxespad=0)
plt.tight_layout()
plt.savefig("figs/pitcount-bySite.png", format='png',dpi=150)
```
